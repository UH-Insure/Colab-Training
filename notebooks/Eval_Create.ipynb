```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from dotenv import load_dotenv
```

```python
VERSION = "v1.1s"
VARIANT = "nocomments"
#data/holdout_message_format_slices_v1.1s.jsonl
holdout_messages_df = pd.read_json(f"data/holdout_message_format_{VERSION}.jsonl", lines=True)
holdout_messages_df.head()
```

```python
eval_tests = []
for idx, row in holdout_messages_df.iterrows():
    messages = row['messages']
    task = ""
    for message in messages:
        if message['role'] == 'user':
            task += message['content'] + "\n"
    test = {
        "task_id": idx + 1,
        "type": "function",
        "task": task,
        "test_setup_code": "",
        "test_list": "",
        "challenge_test_list": ""
    }
    eval_tests.append(test)
eval_tests_df = pd.DataFrame(eval_tests)
eval_tests_df.head()
```
